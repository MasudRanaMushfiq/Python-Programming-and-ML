Loops in Python
What is Loop?
A loop is a control structure used to execute a block of code repeatedly
Which avoids redundancy and reduces code length.
Example: Office commute, Railway weekly Schedule etc.

Types of Loops in Python
For Loop: keyword is for
Syntax: for item in iterable:
It can iterate over sequences (e.g., lists, tuples, strings, ranges). Example: for item in myList:
Example:
for i in range(5):
print("Iteration:", i)

while Loop
Syntax: while condition:
Executes as long as the condition is true.
There are 3 parts in a loop

Example:
count = 0
while count < 5:
    print("Count:", count)
    count += 1

Control Statements
break
Stops the loop immediately.
Example:
for i in range(10):
    if i == 5:
        break
    print(i)

continue
Skips the current iteration and moves to the next.
Example:
for i in range(5):
    if i == 2:
        continue
    print(i)
pass
A placeholder to maintain loop structure.
Example:
for i in range(5):
    if i == 3:
        pass
    print(i)
Nested Loops
Definition: Loops inside another loop.
Example:
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")

Infinite Loops
What happens when a loop runs indefinitely?
Example of an infinite loop:
while True:
    print("This will run forever unless stopped!")
Ways to avoid or break out of infinite loops.

```python
# What is a loop?
# A loop is a control structure that allows a block of code to be executed repeatedly.
# This is useful for performing the same action multiple times without writing the same code over and over.
# It helps to avoid redundancy and reduce the length of your code.

# Example of a simple loop using 'for'
# We will print the message "Hello, Loops!" 5 times.
for i in range(5):
  print("Hello, Loops!")

# Without a loop, we would have to write print("Hello, Loops!") five separate times.
# This becomes very inefficient if we need to repeat an action many times.
```

```python
# For Loop
# Syntax: for item in iterable:
# The 'for' loop iterates over the items of any sequence (such as a list, tuple, string, or range)
# in the order that they appear in the sequence.

# Example: Iterating over a list of fruits
fruits = ["apple", "banana", "cherry"]

# The loop will assign each element of the 'fruits' list to the variable 'fruit' in turn.
# For each 'fruit' in the 'fruits' list, the code inside the loop will be executed.
for fruit in fruits:
  # This line prints the current fruit.
  print(fruit)

# After the loop finishes, all items in the 'fruits' list will have been processed.
```

```python
# while Loop
# Syntax: while condition:
# The 'while' loop executes a block of code as long as a specified condition is true.

# Initialization: This sets up the loop control variable before the loop starts.
count = 0

# Condition: This is checked at the beginning of each loop iteration.
# The loop continues to execute as long as 'count' is less than 5.
while count < 5:
  # Code inside the loop
  print("Count:", count)

  # Increment/Decrement: This changes the loop control variable.
  # It's crucial for ensuring the condition eventually becomes false,
  # preventing an infinite loop. Here, we increment 'count'.
  count += 1

# The loop terminates when 'count' is no longer less than 5.
```

```python
# break statement
# The 'break' statement is used to exit a loop prematurely.
# When 'break' is encountered inside a loop, the loop is immediately terminated,
# and the program execution continues with the first statement after the loop.

print("Starting the loop...")

# Example: Using break to stop a loop when a specific condition is met
for number in range(10):
  # Print the current number in the loop
  print(f"Current number: {number}")

  # Check if the current number is 5
  if number == 5:
    print("Condition met: number is 5.")
    # Use break to exit the loop immediately
    break
  # This line will not be executed if break is triggered

# This print statement is outside the loop.
# It will be executed after the loop finishes, either by completing all iterations
# or by being exited early due to the 'break' statement.
print("Loop finished.")
print("The loop was stopped when the number reached 5 due to the 'break' statement.")
```

```python
# continue statement
# The 'continue' statement is used to skip the rest of the code inside a loop
# for the current iteration only.
# When 'continue' is encountered, the loop's execution jumps to the next iteration
# (for a 'for' loop) or re-evaluates the condition (for a 'while' loop).

print("Starting the loop...")

# Example: Using continue to skip printing even numbers
# We will iterate from 0 to 9.
for num in range(10):
  print(f"Checking number: {num}")

  # Check if the current number is even
  if num % 2 == 0:
    print(f"Number {num} is even. Skipping the rest of this iteration...")
    # Use continue to skip the rest of the code in this iteration
    # and move to the next number in the range.
    continue

  # This print statement will only be executed if the number is odd,
  # because the 'continue' statement is skipped for odd numbers.
  print(f"Number {num} is odd. Printing the number: {num}")

# This print statement is outside the loop and will be executed after the loop finishes.
print("Loop finished.")
print("Only odd numbers were printed because even numbers were skipped using 'continue'.")
```

```python
# pass statement
# The 'pass' statement is a null operation; it does nothing.
# It is used as a placeholder when a statement is syntactically required but you don't want
# any code to execute. This is often used when you are writing code
# and haven't decided what to put in a specific block yet, but need valid syntax.

print("Starting the loop...")

# Example: Using pass as a placeholder in an if condition within a loop
# We will iterate from 0 to 4.
for i in range(5):
  print(f"Current iteration: {i}")

  # Check if the current number is 2
  if i == 2:
    # If the number is 2, we don't want to do anything specific for now.
    # 'pass' serves as a placeholder to keep the syntax correct.
    print("Condition met: i is 2. Using 'pass' as a placeholder.")
    pass # Do nothing when i is 2
    # You could add code here later if needed.

  # This block executes for all numbers, including when i is 2,
  # because 'pass' does not skip the rest of the loop iteration.
  print(f"Processing number {i} after the if/pass block.")

# This print statement is outside the loop.
print("Loop finished.")
print("The 'pass' statement did not affect the loop's execution flow.")
```

```python
# Nested Loops
# A nested loop is a loop inside another loop.
# The inner loop completes all of its iterations for each single iteration of the outer loop.

print("Starting nested loops...")

# Outer loop: This loop will iterate a specified number of times.
# For each iteration of the outer loop, the entire inner loop will run.
for i in range(3):
  print(f"\n--- Outer loop iteration: {i} ---")
  # The code inside the outer loop is executed here.

  # Inner loop: This loop is nested inside the outer loop.
  # It will iterate a specified number of times for *each* value of 'i' in the outer loop.
  for j in range(2):
    # The code inside the inner loop is executed for each combination of i and j.
    print(f"  Inner loop iteration: i={i}, j={j}")
    # This print statement shows that the inner loop runs completely for each outer loop iteration.

  # This line is inside the outer loop but outside the inner loop.
  # It is executed after the inner loop finishes for the current outer loop iteration.
  print(f"--- Inner loop finished for outer iteration {i} ---")

# This print statement is outside both loops and is executed after the outer loop finishes.
print("\nNested loops finished.")
print("The inner loop ran 2 times for each of the 3 outer loop iterations, for a total of 3 * 2 = 6 inner loop executions.")
```

```python
# Infinite Loops
# An infinite loop is a loop that runs indefinitely because its condition never becomes false.
# This can happen if the loop control variable is not updated, or if the condition is always true.
# Infinite loops can cause programs to hang or consume excessive resources.

print("Demonstrating an infinite loop (with a controlled break)...")

# Example of an infinite loop using 'while True:'
# The condition 'True' is always true, so this loop would normally run forever.
# To avoid this in real code, always ensure your loop condition will eventually be false.
count = 0
while True:
  # Print the current count
  print(f"Iteration: {count}")

  # This part demonstrates how to manually stop the loop for this example.
  # In a true infinite loop scenario without this check, this loop would continue indefinitely.
  count += 1
  if count >= 5:
    print("Reached the break limit for demonstration. Stopping the loop.")
    break # Exit the loop after 5 iterations

  # In a real infinite loop, you would not have a condition like 'if count >= 5'
  # that eventually becomes true to break out of the loop within the code itself.

# How to avoid infinite loops:
# - Ensure the condition used in the 'while' loop will eventually evaluate to False.
# - Make sure you update the variables used in the loop condition within the loop body.
#   For example, in a 'while count < 10:' loop, make sure 'count' is incremented or changed.
# - Carefully review your loop logic and conditions.

# How to stop an infinite loop when a program is running:
# In most programming environments (like a terminal or Jupyter Notebook), you can typically
# press Ctrl+C (or Command+C on macOS) to send an interrupt signal to stop the execution.

print("Loop execution finished.")
```

**Reasoning**:
Implement the steps to calculate and print the average of numbers divisible by 3 and less than 100 using a loop.



```python
# 1. Initialize variables for sum and count.
sum_divisible_by_3 = 0
count_divisible_by_3 = 0

# 2. Use a loop to iterate through numbers from 1 to 99.
# range(1, 100) generates numbers from 1 up to (but not including) 100.
for number in range(1, 100):
  # 3. Check if the current number is divisible by 3.
  if number % 3 == 0:
    # 4. If divisible by 3, add it to the sum and increment the count.
    sum_divisible_by_3 += number
    count_divisible_by_3 += 1

# 5. Check if the count is greater than zero to avoid division by zero.
if count_divisible_by_3 > 0:
  # 6. Calculate the average.
  average = sum_divisible_by_3 / count_divisible_by_3
  # 7. Print the average.
  print(f"The average of numbers divisible by 3 and less than 100 is: {average}")
else:
  print("No numbers divisible by 3 found in the specified range.")
```

**Reasoning**:
The subtask is to write a program to check if a number is prime. This involves getting input from the user, handling edge cases (numbers <= 1), and iterating through potential divisors up to the square root of the number. I will combine these steps into a single code block.



```python
import math

# 1. Prompt the user to enter an integer and store it in a variable.
try:
  num = int(input("Enter an integer to check if it's a prime number: "))

  # 2. Check if the number is less than or equal to 1. If it is, print that it is not a prime number.
  if num <= 1:
    print(f"{num} is not a prime number.")
  # 3. If the number is greater than 1, iterate from 2 up to the square root of the number (inclusive).
  else:
    is_prime = True # Assume the number is prime initially
    # Iterate from 2 up to the integer part of the square root of num
    for i in range(2, int(math.sqrt(num)) + 1):
      # 4. Inside the loop, check if the number is divisible by the current iteration number.
      if num % i == 0:
        # 5. If the number is divisible by any number in the loop, print that it is not a prime number and break out of the loop.
        print(f"{num} is not a prime number (it is divisible by {i}).")
        is_prime = False
        break # No need to check further if a divisor is found

    # 6. If the loop completes without finding any divisors, print that the number is a prime number.
    if is_prime:
      print(f"{num} is a prime number.")

except ValueError:
  print("Invalid input. Please enter a valid integer.")
```

```python
# 1. Prompt the user to enter a non-negative integer and store it in a variable.
# Include error handling to ensure the input is a valid integer.
try:
  num = int(input("Enter a non-negative integer to calculate its factorial: "))

  # 2. Initialize a variable to 1 to store the factorial result.
  factorial = 1

  # 3. If the entered number is 0, print that the factorial is 1.
  if num == 0:
    print(f"The factorial of {num} is 1.")

  # 4. If the entered number is negative, print an appropriate message.
  elif num < 0:
    print("Factorial is not defined for negative numbers.")

  # 5. If the entered number is positive, use a for loop.
  else:
    # Iterate from 1 up to the entered number (inclusive).
    # range(1, num + 1) generates numbers from 1 up to and including num.
    for i in range(1, num + 1):
      # 6. Inside the loop, multiply the current number with the factorial result.
      # and update the factorial result variable.
      factorial *= i

    # 7. After the loop finishes, print the calculated factorial.
    print(f"The factorial of {num} is {factorial}.")

except ValueError:
  print("Invalid input. Please enter a valid integer.")
```

Implement the prime number finding logic using nested loops as described in the instructions.



```python
import math

# 1. Iterate through numbers from 2 to 100 (inclusive).
print("Prime numbers between 1 and 100 are:")
for num in range(2, 101):
  # 2. For each number, assume it is prime initially.
  is_prime = True

  # 3. Check for divisibility by numbers from 2 up to the square root of the current number.
  # We only need to check up to the square root because if a number 'num' has a divisor
  # greater than its square root, it must also have a divisor less than its square root.
  # The range goes from 2 up to the integer part of the square root of num + 1
  # to include the square root itself if it's an integer.
  for i in range(2, int(math.sqrt(num)) + 1):
    # 4. If the number is divisible by any number in this range, mark it as not prime and break the inner loop.
    if num % i == 0:
      is_prime = False
      break # No need to check further divisors

  # 5. If the inner loop completes without finding any divisors, print the number as it is prime.
  if is_prime:
    print(num)
```
