```python
# @title Default title text
# variable
num = 10
print(num)

#Type check
a = "Man"
print(type(a))

#composite array
my_mixed_array = [1, "hello", 3.14, True, [1, 2, 3]]
print(my_mixed_array)

#complex number
z = complex(3, 4)
print(z)

# Tuple to store a pair of values like RGB color or coordinate, more memory efficient, not changable(constannt).
coordinates = (10, 20)

print(coordinates[0], coordinates[1])

```

```python
first = 4
second = 3

#Exponent operator
third = first ** second
print(third)

#Floor Division operator
fourth = first // second
print(fourth)
```

```python
x = "Bangladesh"
y = "Bangla"

#Membership operator
z = y in x
print(z)

m = y not in x
print(m)
```

```python
a = "Man"
b = 20

#Identity operator
c = a is b
print(c)

d = a is not b
print(d)
```

```python
# Same name in different scope is allowed.
x = 100
def test():
    x = 200  # shadows the global x
    print(x)

test()
print(x)  # global x remains unchanged


# x = 5
# def bad_func():
#   x += 1  # Error! Python thinks x is local, but it's referenced before assignment

# bad_func()


y = 5
def good_func():
    global y
    y += 1
    print(y)

good_func()
```

```python
# Dictionary (Collection of Key Value pair)
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
print(person["name"])

# Set (An unordered collection of unique values.)
fruits = {"apple", "banana", "cherry", "apple"}
print(fruits)

```

```python
# List (Mixed type, index accessing, mutable, duplicate allowed data type)
colors = [ 10, "red", "green", "blue", 29.5]
print(colors[0])
colors[1] = "yellow"
colors.append("purple")
colors.remove("green")
print(colors)

# Map (Applies a function to every item in an iterable.)
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)
```
