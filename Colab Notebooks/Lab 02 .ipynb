```python
# 1. String Concatenation using the '+' operator
# This method joins two or more strings together.
name = "Alice"
greeting_concat = "Hello, " + name + "!"
print(greeting_concat)

# 2. String Formatting using the .format() method
# This method uses placeholders ({}) in the string and replaces them with the values provided to the format() method.
age = 30
greeting_format = "Hello, {}! You are {} years old.".format(name, age)
print(greeting_format)

# 3. String Formatting using f-strings (Formatted String Literals)
# This is a more concise and readable way to embed expressions inside string literals. Precede the string with 'f' or 'F'.
city = "New York"
greeting_fstring = f"Hello, {name}! You live in {city}."
print(greeting_fstring)

# 4. Formatting Numbers within a string
# You can format numbers to a specific number of decimal places using formatting specifiers within the placeholders.
pi = 3.14159265359
# :.2f formats the number to 2 decimal places (f for float).
pi_formatted = f"The value of pi is approximately {pi:.2f}"
print(pi_formatted)

price = 19.9931
# :.2f is commonly used for currency formatting.
price_formatted = f"The price is ${price:.2f}"
print(price_formatted)
```

```python
# Basic if-else statement
# Checks if a condition is true and executes one block, otherwise executes the else block.
temperature = 25
if temperature > 30:
    print("It's hot outside.")
else:
    print("It's not hot outside.")

# Using the 'and' logical operator
# Both conditions must be true for the block to execute.
age = 25
income = 40000
if age >= 18 and income > 30000:
    print("You are eligible for a loan.")
else:
    print("You are not eligible for a loan based on age and income.")

# Using the 'or' logical operator
# At least one condition must be true for the block to execute.
is_student = True
has_scholarship = False
if is_student or has_scholarship:
    print("You qualify for a student discount.")
else:
    print("You do not qualify for a student discount.")

# Using the 'not' logical operator
# Reverses the boolean value of a condition.
is_raining = False
if not is_raining:
    print("It's a good day to go for a walk.")
else:
    print("Stay indoors, it's raining.")

# Nested if statement
# An if statement inside another if statement.
score = 85
if score >= 50:
    print("Pass")
    if score >= 75:
        print("Distinction!")
    else:
        print("Just a pass.")
else:
    print("Fail")

# Using elif for multiple conditions
# Checks conditions sequentially. The first condition that is true executes its block.
grade = 78
if grade >= 90:
    print("Grade: A")
elif grade >= 80:
    print("Grade: B")
elif grade >= 70:
    print("Grade: C")
elif grade >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
# Even/Odd Exercise
# This program checks if a number is even or odd using the modulo operator.
num = 10
if num % 2 == 0:
    print(f"{num} is an even number.")
else:
    print(f"{num} is an odd number.")

# Positive/Negative/Zero Exercise
# This program determines if a number is positive, negative, or zero.
num = -5
if num > 0:
    print(f"{num} is a positive number.")
elif num < 0:
    print(f"{num} is a negative number.")
else:
    print(f"{num} is zero.")

# Calculator Exercise
# This program performs basic arithmetic operations based on user input (simulated here).
num1 = 10
num2 = 5
operator = "+" # Can be "+", "-", "*", or  "/"

if operator == "+":
    result = num1 + num2
    print(f"{num1} {operator} {num2} = {result}")
elif operator == "-":
    result = num1 - num2
    print(f"{num1} {operator} {num2} = {result}")
elif operator == "*":
    result = num1 * num2
    print(f"{num1} {operator} {num2} = {result}")
elif operator == "/":
    # Check for division by zero
    if num2 != 0:
        result = num1 / num2
        print(f"{num1} {operator} {num2} = {result}")
    else:
        print("Error: Division by zero.")
else:
    print("Invalid operator.")

# Grading System Exercise
# This program assigns a grade based on a given mark.
mark = 85

if mark >= 90:
    print("Grade: A+")
elif mark >= 80:
    print("Grade: A")
elif mark >= 75:
    print("Grade: A-")
elif mark >= 70:
    print("Grade: B")
elif mark >= 60:
    print("Grade: C")
elif mark >= 50:
    print("Grade: D")
else:
    print("Grade: F")

# Triangle Type Checker Exercise
# This program determines if a triangle is equilateral, isosceles, or scalene based on its side lengths.
side1 = 5
side2 = 5
side3 = 7

if side1 == side2 == side3:
    print("The triangle is equilateral.")
elif side1 == side2 or side1 == side3 or side2 == side3:
    print("The triangle is isosceles.")
else:
    print("The triangle is scalene.")

# Percentage to Decimal Exercise
# This program converts a percentage value to its decimal equivalent.
percentage = 125
decimal_value = percentage / 100
print(f"{percentage}% is equivalent to {decimal_value}")

# Centigrade to Fahrenheit Exercise
# This program converts a temperature from Celsius to Fahrenheit.
celsius = 25
fahrenheit = (celsius * 1.8) + 32
print(f"{celsius}°C is equal to {fahrenheit}°F")

# Inches to Feet/Inches Exercise
# This program converts a total number of inches into feet and remaining inches.
total_inches = 185
feet = total_inches // 12  # Integer division for feet
remaining_inches = total_inches % 12 # Modulo for remaining inches
print(f"{total_inches} inches is {feet} feet and {remaining_inches} inches.")

# Copy Center Cost Exercise
# This program calculates the cost of copies based on a tiered pricing structure.
num_copies = 150
cost_per_copy_first_100 = 50
cost_per_copy_additional = 30
first_100_copies_limit = 100

if num_copies <= first_100_copies_limit:
    total_cost = num_copies * cost_per_copy_first_100
else:
    cost_first_100 = first_100_copies_limit * cost_per_copy_first_100
    additional_copies = num_copies - first_100_copies_limit
    cost_additional = additional_copies * cost_per_copy_additional
    total_cost = cost_first_100 + cost_additional

print(f"The total cost for {num_copies} copies is {total_cost} won.")

# Leap Year Exercise
# This program checks if a given year is a leap year according to the Gregorian calendar rules.
year = 2000

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
import random
yourguess = input("Enter Your Guess (rock, paper, or scissors): ").lower()

randnum = random.randint(1, 3)

# Determine and print the computer's choice based on the random number
print("Computer chose:")
if randnum == 1:
    print("Rock")
    computer_choice = "rock" # Store computer's choice as a string
elif randnum == 2:
    print("Paper")
    computer_choice = "paper" # Store computer's choice as a string
elif randnum == 3:
    print("Scissors")
    computer_choice = "scissors" # Store computer's choice as a string

# Determine the winner based on the rules of Rock Paper Scissors
if yourguess == computer_choice:
    print("It's a tie!")
elif (yourguess == "rock" and computer_choice == "scissors") or \
     (yourguess == "paper" and computer_choice == "rock") or \
     (yourguess == "scissors" and computer_choice == "paper"):
    print("You win!")
else:
    print("You lose!")
```
