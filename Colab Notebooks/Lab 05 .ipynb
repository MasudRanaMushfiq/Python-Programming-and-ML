```python
# Built-in String Functions

# 1. len()
# Find the length of a string.
text_len = input("Enter a string to find its length: ")
print("Length of the string:", len(text_len)) # The len() function returns the number of characters in the string.

# 2. split() and join()
# Split a string into words and join them back.
sentence_split_join = input("Enter a sentence to split and join: ")
words = sentence_split_join.split() # The split() method splits the string into a list of words using whitespace as a delimiter by default.
print("List of words:", words)
joined_sentence = " ".join(words) # The join() method concatenates the elements of an iterable (like a list of words) into a single string, using the string it's called on as the separator.
print("Joined back:", joined_sentence)

# 3. startswith() and endswith()
# Check if a string starts or ends with a specific substring.
text_start_end = input("Enter a string to check start/end: ")
print("Starts with 'Hello':", text_start_end.startswith("Hello")) # The startswith() method returns True if the string starts with the specified substring, otherwise False.
print("Ends with 'world':", text_start_end.endswith("world")) # The endswith() method returns True if the string ends with the specified substring, otherwise False.

# 4. count()
# Count occurrences of a specific character or substring.
text_count = input("Enter a string to count characters: ")
char_to_count = input("Enter a character to count: ")
print(f"The character '{char_to_count}' occurs {text_count.count(char_to_count)} times.") # The count() method returns the number of occurrences of a substring within the string.

# 5. find() and index()
# Locate a substring in a string.
text_find_index = input("Enter a string to find a word: ")
word_to_find = input("Enter a word to find: ")
print(f"'{word_to_find}' found at index using find():", text_find_index.find(word_to_find)) # The find() method returns the lowest index in the string where the substring is found. Returns -1 if not found.
try:
    print(f"'{word_to_find}' found at index using index():", text_find_index.index(word_to_find)) # The index() method is similar to find(), but raises a ValueError if the substring is not found.
except ValueError:
    print(f"'{word_to_find}' not found using index()")

# 6. isalpha(), isdigit(), isalnum()
# Validate string content.
text_validate = input("Enter a string to validate: ")
print("Contains only letters (isalpha()):", text_validate.isalpha()) # isalpha() returns True if all characters in the string are alphabetic and there is at least one character, False otherwise.
print("Contains only digits (isdigit()):", text_validate.isdigit()) # isdigit() returns True if all characters in the string are digits and there is at least one character, False otherwise.
print("Contains only alphanumeric characters (isalnum()):", text_validate.isalnum()) # isalnum() returns True if all characters in the string are alphanumeric (letters or digits) and there is at least one character, False otherwise.

# 7. capitalize(), title(), and swapcase()
# Modify string case styles.
text_case = input("Enter a sentence to change case: ")
print("Capitalized:", text_case.capitalize()) # capitalize() returns a copy of the string with its first character capitalized and the rest lowercased.
print("Title Case:", text_case.title()) # title() returns a titlecased version of the string, where words start with an uppercase character and the remaining characters are lowercase.
print("Swap Case:", text_case.swapcase()) # swapcase() returns a copy of the string with uppercase characters converted to lowercase and vice versa.
```

Provide the Python code solutions for the Palindrome Checker, Word Frequency Counter, and Anagram Checker challenges with comments explaining the logic for each.



```python
# Challenges

# 1. Palindrome Checker
# Write a program to check if a given string is a palindrome.
text_palindrome = input("Enter a string to check if it's a palindrome: ")
# To check for a palindrome, compare the original string with its reversed version.
# Slicing with [::-1] creates a reversed copy of the string.
if text_palindrome == text_palindrome[::-1]:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")

# 2. Word Frequency Counter
# Count the frequency of each word in a given sentence.
sentence_frequency = input("Enter a sentence to count word frequency: ")
# Split the sentence into a list of words. The split() method handles whitespace.
words_list = sentence_frequency.split()
# Use a dictionary to store word frequencies.
# Iterate through the list of words. For each word, increment its count in the dictionary.
# The get() method with a default value of 0 is useful for handling the first occurrence of a word.
frequency = {}
for word in words_list:
    # Convert word to lowercase to treat "Word" and "word" as the same.
    word_lower = word.lower()
    frequency[word_lower] = frequency.get(word_lower, 0) + 1
print("Word frequencies:", frequency)

# 3. Anagram Checker
# Check if two strings are anagrams of each other.
str1_anagram = input("Enter the first string to check for anagrams: ")
str2_anagram = input("Enter the second string to check for anagrams: ")
# Two strings are anagrams if they contain the same characters with the same frequencies.
# A simple way to check this is to sort the characters of both strings and compare the sorted results.
# Convert strings to lowercase and remove spaces before sorting for a more robust check.
str1_cleaned = "".join(sorted(str1_anagram.lower().replace(" ", "")))
str2_cleaned = "".join(sorted(str2_anagram.lower().replace(" ", "")))

# Compare the cleaned and sorted strings.
if str1_cleaned == str2_cleaned:
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```
